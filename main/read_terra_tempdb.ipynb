```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

temp = pd.read_csv('/Users/nilslec/Downloads/influxdata_2024-10-23T07_26_58Z.csv',names = ['errState','PWMValue', 'SP', 'temp', 'time'], header = 3, dtype = {'errState': bool, 'relState': bool, 'SP': np.float64, 'temp': np.float64},parse_dates=['time'])

print(temp.shape)
# Drop rows with NaN index
temp.dropna(subset=['time'], inplace=True)

# Set the 'time' column as the index
temp.set_index('time', inplace=True)

temp = temp.sort_values('time')

print(temp.shape)
print(temp)
# Plot the temperature over time
plt.figure(figsize=(10, 6))
plt.plot(temp.index, temp['temp'])
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.title('Temperature Over Time')
plt.grid(True)
plt.show()


plt.figure(figsize=(10, 6))
plt.scatter(temp.index, temp['PWMValue'])
plt.xlabel('Time')
plt.ylabel('PWM Value')
plt.title('PWM Value')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(temp.index, temp['SP'])
plt.xlabel('Time')
plt.ylabel('SetPoint')
plt.title('Temp Set Point')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(temp.index, temp['errState'])
plt.xlabel('Time')
plt.ylabel('Error State')
plt.title('Error State')
plt.grid(True)
plt.show()
```
